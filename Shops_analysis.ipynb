```python
def online_func(x):
    if "Digital" in x: 
        return 1
    if "Internet" in x:
        return 1
    return 0 

shop_df['online'] = shop_df['shop_name'].apply(online_func)
shop_df
```

```python
def name_split(x):
    xs = x.split(" ")
    xs[0].strip()
    if xs[0] == "!" or xs[0] == "":
        if xs[1] == "St.":
            name = "St. " + xs[2]
            return name 
        return xs[1]
    return xs[0]

shop_df['city'] = shop_df['shop_name'].apply(name_split)
shop_df
```

```python
just_cities = shop_df['city']
```

```python
just_cities.drop_duplicates(inplace=True)
just_cities.drop(12, inplace=True)
just_cities.drop(9, inplace=True)
just_cities.head()
```

```python
geolocator = Nominatim(user_agent='myapplication')
locations = []
for c in just_cities.values:
    location = geolocator.geocode(c)
    if location is not None:
        loc = location[-1]
        locations.append(loc)
```

```python
just_cities['locations'] = locations
just_cities
```
