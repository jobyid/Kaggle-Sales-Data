```python
import pandas as pd 
import numpy as np 

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cat_df  = pd.read_csv('eng_categories.csv')
item_df = pd.read_csv('eng_items.csv')
```

```python
cats         = []
sub_cats     = []
sub_sub_cats = []
def split_catergories(x):
    xs = x.split(" ")
    w = []
    for x in xs:
        x.strip()
        if x != "":
            w.append(x)
    cat = w[0]
    sub_cat = ""
    sub_sub_cat = ""
    if len(w) > 1:
        if w[1] != "-":
            sub_cat = w[1]
        elif w[1] == "-":
            sub_cat = w[2]
        if len(w) > 4:
            sub_sub_cat = w[3] + " " + w[4]
        elif len(w) > 3:
            sub_sub_cat = w[3]
    
    #print(cat," ", sub_cat, " ", sub_sub_cat)
    cats.append(cat)
    sub_cats.append(sub_cat)
    sub_sub_cats.append(sub_sub_cat)
    
cat_df.category_name.apply(split_catergories)
cat_df['cat']         = cats
cat_df["sub_cat"]     = sub_cats
cat_df["sub_sub_cat"] = sub_sub_cats
cat_df.head()
```

```python
cv1 = CountVectorizer()
xcv = cv1.fit_transform(item_df.item_name)
tokens = cv1.get_feature_names()
item_df_cv1 = pd.DataFrame(data=xcv.toarray(), columns=tokens)
item_df_cv1.head()
```

```python
cv2 = CountVectorizer(ngram_range=(1,2))
xcv2 = cv2.fit_transform(item_df.item_name)
tokens = cv2.get_feature_names()
item_df_cv2 = pd.DataFrame(data=xcv2.toarray(), columns=tokens)
item_df_cv2.head()
```

```python
tf1 = TfidfVectorizer()
xtf = tf1.fit_transform(item_df.item_name)
tokens = tf1.get_feature_names()
item_df_tf1 = pd.DataFrame(data=xtf.toarray(), columns=tokens)
item_df_tf1.head()
```

```python
tf2 = TfidfVectorizer(ngram_range=(1,2))
xtf2 = tf2.fit_transform(item_df.item_name)
tokens = tf2.get_feature_names()
item_df_tf2 = pd.DataFrame(data=xtf2.toarray(), columns=tokens)
item_df_tf2.head()
```

```python
cv2m = CountVectorizer(ngram_range=(1,2), min_df=3)
xcv2m = cv2m.fit_transform(item_df.item_name)
tokens = cv2m.get_feature_names()
item_df_cv2m = pd.DataFrame(data=xcv2m.toarray(), columns=tokens)
item_df_cv2m.head()
```
